```python
#税抜価格を入力し整数に変換する
kakaku = int(input("税抜価格を入力してください："))

#税込価格(10%)とする。尚、小数点以下を四捨五入する
kakaku_tax = int(round(kakaku*1.1))

#送料無料を初期値として設定する
souryou = 0

#計算された税込価格によって処理を分岐させる
if kakaku_tax >= 2000:
    #税込価格2000円以上の場合
    print("送料が無料です")
    #送料は無料のまま

#税込価格が2000円未満の場合
else:
    print("送料として350円かかります")
    #送料350円を税込価格にプラスする
    souryou = 350

#送料込の価格を計算する
final_price = kakaku_tax + souryou

#送料込価格を表示する
print(f"送料込価格は{final_price}円です。")
```
