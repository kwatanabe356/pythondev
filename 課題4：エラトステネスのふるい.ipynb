```python
#関数作成
def is_sosuu(n):
    #1は素数ではない
    if n == 1:
        return False
    #2,3,5,7は素数
    if n in (2,3,5,7):
        return True
    #2,3,5,7を除いた倍数は素数ではない
    if n % 2 == 0 or n % 3 == 0 or n % 5 == 0 or n % 7 == 0:
        return False
    #それ以外の数は、1からその数の平方根までで割り切れるかどうか調べる
    for i in range(11, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True
#1から100までの素数を表示
#1から100までのリストを作成
for i in range(1,101):
    if is_sosuu(i):
        print(i)
```
