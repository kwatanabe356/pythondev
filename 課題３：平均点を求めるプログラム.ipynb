```python
#合計点数
total_score = 0
#人数
people = 0
#無限ループ
while True:
    #点数の入力を整数で求める
    score = int(input("点数を入力してください："))
    #終了条件：－１が入力されたらループ終了
    if score == -1:
        break
    #入力された点数を合計に加算し、人数をカウント
    total_score += score
    people += 1
#結果表示
if people > 0:
   #平均点計算
   average_score = total_score / people
   #平均点を四捨五入（小数点第２位で）、小数点第１位迄表示
   print(f"{people}人のテスト平均点は{average_score:.1f}点です。")
else:
   print("テストの点数が入力されませんでした")

```
